```python
import numpy as np
import pandas as pd

pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

sns.set_style('whitegrid')
import matplotlib.font_manager as fm
plt.rc('font', family=fm.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()) 
```

```python
mz=pd.read_csv('MZ세대를 위한 신용카드 수요조사의 사본.csv')
```

```python
plt.figure(figsize=(16,6))
sns.countplot(x=mz['1. 귀하는 학생입니까?'], palette='Reds_r')
```

```python
plt.figure(figsize=(16,6))
sns.countplot(x=mz['1-1. 학생이라면 월 근로소득이 있습니까?'], palette='Reds_r')
```

```python
plt.figure(figsize=(16,6))
sns.countplot(x=mz['2. 월 근로소득이 얼마나 됩니까?'], palette='Reds_r',order=['30만원 이하', '50만원 이하', '100만원 이하', '200만원 이하', '200만원 초과'])
plt.title('월 근로소득')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.countplot(x=mz['3. 한달에 지출이 얼마입니까?'], palette='Reds_r',order=['30만원 이하', '50만원 이하', '100만원 이하', '200만원 이하', '200만원 초과'])
plt.title('월 지출')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.countplot(x=mz['4. 귀하는 자기 명의로 된 신용카드가 있습니까?'], palette='Reds_r')
plt.title('4. 귀하는 자기 명의로 된 신용카드가 있습니까?')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.countplot(x=mz['4-1. 신용카드가 있다면 한달에 얼마씁니까?'], palette='Reds_r',order=['30만원 이하', '50만원 이하', '100만원 이하', '200만원 이하', '200만원 초과'])
plt.title('신용카드 지출')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.countplot(x=mz['4-2. 없다면 이유는 무엇입니까?'], palette='Reds_r',order=mz['4-2. 없다면 이유는 무엇입니까?'].value_counts().index)
plt.title('신용카드 없는 이유')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.countplot(x=mz['5. 귀하는 대출을 받은 적 있습니까?'], palette='Reds_r')
plt.title('대출 유무')
plt.show()
```

```python
aa=mz['5-1. 있다면 어떤 대출을 받았습니까?']
aa=pd.DataFrame(aa)
aa=aa.loc[:29]
```

```python
aa.loc[:12,'5-1. 있다면 어떤 대출을 받았습니까?']='전월세 대출'
aa.loc[13:23,'5-1. 있다면 어떤 대출을 받았습니까?']='학자금 대출'
aa.loc[24:28,'5-1. 있다면 어떤 대출을 받았습니까?']='생활비 대출'
aa.loc[29,'5-1. 있다면 어떤 대출을 받았습니까?']='신용 대출'
```

```python
plt.figure(figsize=(16,6))
sns.countplot(x=aa['5-1. 있다면 어떤 대출을 받았습니까?'], palette='Reds_r')
plt.title('대출 종류')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.countplot(x=mz['5-2. 없다면 이유는 무엇입니까?'], palette='Reds_r',order=mz['5-2. 없다면 이유는 무엇입니까?'].value_counts().index)
plt.title('대출을 하지않는 사유')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.countplot(x=mz['6. 귀하는 MZ세대(금융 소외계층) 맞춤형 금융 상품이 있다면 사용하실 의향이 있습니까?'], palette='Reds_r',order=mz['6. 귀하는 MZ세대(금융 소외계층) 맞춤형 금융 상품이 있다면 사용하실 의향이 있습니까?'].value_counts().index)
plt.title('맞춤금융상품 선호도')
plt.show()
```
